```python
df.dropna(subset=['Price'], inplace=True)
```

```python
X = df.drop('Price', axis=1)
y = df['Price'].copy()
```

```python
df
```

```python
X['Doors'].fillna(X['Doors'].median())      # Replaces NaN with the median
```

```python
X['Make'].fillna(value='Missing').value_counts()
```

```python
X['Make'] = X['Make'].fillna(X['Make'].mode()[0])
```

```python
X['Make'].mode()
```

```python
X['Colour'] = X['Colour'].fillna(X['Colour'].mode()[0])
```

```python
X['Colour'].mode()
```

```python
X.isna().sum()
```

```python
X.Make.value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_enc = OrdinalEncoder()
```

```python
X['Make_encoded'] = ordinal_enc.fit_transform(pd.DataFrame(X['Make']))
```

```python
X
```

```python
X['Colour_encoded'] = ordinal_enc.fit_transform(pd.DataFrame(X['Colour']))
```

```python
X
```

```python
X.drop(['Make', 'Colour'], axis=1, inplace=True)
```

```python
X
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X, y)
```

```python
rfr.score(X, y)
```
